```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
```

```python
#1. Write a Python program to test if a given page is found or not on the server.
#2. Write a Python program to extract and display all the header tags from en.wikipedia.org/wiki/Main_Page
url = "https://en.wikipedia.org/wiki/Main_Page?"
test = requests.get(url)
print(test.reason)
html_data = requests.get(url).text
```

```python
soup = BeautifulSoup(html_data)
```

```python
print(soup.prettify())
```

```python
headers = ["h1", "h2", "h3", "h4", "h5", "h6"]
rows = soup.find_all(headers)
for row in rows:
    print(row)
```

```python
url = "https://en.wikipedia.org/wiki/Peter_Jeffrey_(RAAF_officer)"
r = requests.get(url).text
soup = BeautifulSoup(r)
print(soup.prettify())
```

```python
driver = webdriver.Chrome()
driver.get("https://twitter.com/BBCWorld")

content = driver.page_source
driver.close()
```

```python
soup = BeautifulSoup(content)
print(soup.prettify())
```

```python
for row in soup.find_all("span", string = lambda word: word in ("читачів", "читач", "читачі")):
    print(row.find_parent().text)
```

```python
url = "https://en.wikipedia.org/wiki/Encyclopedia"
response = requests.get(url)
print("Status code: ", response.status_code, end = "\n\n")
print("Headers: ", response.headers, end = "\n\n")
print("url: ", response.url, end = "\n\n")
print("History", response.history, end = "\n\n")
print("Encoding", response.encoding, end = "\n\n")
print("Reason: ", response.reason, end = "\n\n")
print("Cookies: ", response.cookies, end = "\n\n")
print("Elapsed: ", response.elapsed, end = "\n\n")
print("Request: ", response.request, end = "\n\n")
print("Content: ", response.content[:100])
```
